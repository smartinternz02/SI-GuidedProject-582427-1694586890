1. Make data fit for model.

### 1. Making data fit for model.

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the scaled features
X_scaled = scaler.fit_transform(X_encoded)

# Convert the scaled features back to a DataFrame (optional)
X_scaled_df = pd.DataFrame(X_scaled, columns=X_encoded.columns)

# Check data
X_scaled_df.head()
```

### 2. Import the model building Libraries

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

### 3. Initializing the model

```python
logistic_model = LogisticRegression(random_state=42)
random_forest_model = RandomForestClassifier(random_state=42)
```

### 4. Training and Testing the Model

```python
# Training and testing the Logistic Regression model
logistic_model.fit(X_train, y_train)
logistic_result = logistic_model.predict(X_test)

# Training and testing the Random Forest model
random_forest_model.fit(X_train, y_train)
random_forest_result = random_forest_model.predict(X_test)
```

### 5. Evaluation

```python
# Evaluation of Logistic Regression model
logistic_accuracy = accuracy_score(y_test, logistic_result)

print("Logistic Regression Model Accuracy:", logistic_accuracy)
```
